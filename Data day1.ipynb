```python
data = np.loadtxt('ratings.txt', delimiter='::', dtype='int')
data
```

```python
print(data.ndim)
print(data.shape)
print(data.size)
```

```python
전체평점 = np.mean(data[:,2])
전체평점
```

```python
user_id = np.unique(data[:,0])   # 중복제거
user_id
```

```python
print(user_id.shape)   #행,열 형태
```

```python
data[:,0]  #전체행 0번 열 전부 표시
```

```python
data[:,0] == 1   #전체행 0번 열 중에서 1번은 참으로 만들기
```

```python
data[data[:,0] == 1]
```

```python
list1 = []

for user in user_id:
    user_mean = data[data[ : , 0] == user]  # 이 항이 바로 위에 것
    user_mean = user_mean[: ,2].mean()  # 2번열만 지정한 후 평균
    list1.append([user, user_mean])
```

```python
list2 = []

for user in user_id:
    user_mean = data[data[ : , 0] == user, 2].mean()   # list 1 이랑 같은 결과
    list2.append([user, user_mean])
```

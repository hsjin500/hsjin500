# 네이버 뮤직 TOP 50 수집 

```python
import requests as req
from bs4 import BeautifulSoup as bs
```

```python
navermusic_list100_url = 'https://music.naver.com/listen/top100.nhn?domain=TOTAL_V2'
```

```python
req.get(navermusic_list100_url)
```

```python
res = req.get(navermusic_list100_url)
res.text
```

```python
naver_music_100list= bs(res.text,'lxml')
```

```python
title = naver_music_100list.select('a._title > span.ellipsis')
singer = naver_music_100list.select('td._artist > a')
```

```python
rank = naver_music_100list.select(f'td.ranking')
len(rank)
```

```python
len(title)
```

```python
len(singer)
```

```python
for num in range(0,len(rank)) :
    print(singer[num].text.strip())
```

```python
for num in range(0,len(rank)) :
    res0 = naver_music_100list.select(f'td.ranking')
    print(f'{rank[num].text}.{title[num].text}  {singer[num].text.strip()}')
```

```python
r_list=[]
for num in range(0,len(rank)) :
    r_list.append(rank[num].text)
print(r_list)
```

```python
t_list=[]
for num in range(0,len(rank)) :
    t_list.append(title[num].text)
print(t_list)
```

```python
s_list=[]
for num in range(0,len(rank)) :
    s_list.append(singer[num].text.strip())
print(s_list)
```

```python
import pandas as pd
```

```python
music_info = {'rank' : r_list, 'title' : t_list, 'singer' : s_list}
pd.DataFrame(music_info)
```

```python
music_info
```

```python
dataframe = pd.DataFrame(music_info)
```

```python
dataframe.set_index('rank', inplace = True )
```

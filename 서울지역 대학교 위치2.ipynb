```python
import pandas as pd
import folium
```

```python
df = pd.read_excel('./서울지역 대학교 위치.xlsx',encoding='euc-kr')
```

```python
seoul_map = folium.Map( location = [37.55,126.98], tiles='Stamen Terrain', zoom_start=12,encoding='utf-8')
```

```python
df.head()
```

```python
for name, lat, lng in zip(df.대학명, df.위도, df.경도):
    folium.Marker([lat,lng],popup=name).add_to(seoul_map)
```

```python
seoul_map.save('./seoul_map.html')
```

# naver에서 광주날씨 검색한 후 온도 크롤링 

```python
##Chrome 드라이버를 실행
driver = wb.Chrome()
## driver를 통해 사이트의 주소를 요청
url = 'https://www.naver.com/'
driver.get(url)
## 검색창 선택
input_search = driver.find_element_by_id('query')
## 검색어 입력
input_search.send_keys('광주날씨')
##엔터
input_search.send_keys(Keys.ENTER)
#온도 데이터 찾기
weather = driver.find_element_by_class_name('todaytemp')
#온도 데이터 출력
print(weather.text)
```

# Google에서 광주날씨 검색한 후 온도 크롤링  

```python
##Chrome 드라이버를 실행
driver = wb.Chrome()
## driver를 통해 사이트의 주소를 요청
url = 'https://www.google.com/'
driver.get(url)
## 검색창 선택
input_search = driver.find_element_by_class_name('gLFyf')
## 검색어 입력
input_search.send_keys('광주날씨')
##엔터
input_search.send_keys(Keys.ENTER)
#온도 데이터 찾기
weather = driver.find_element_by_id('wob_tm')
#온도 데이터 출력
print(weather.text)
```

### Beautiful soup 이랑 같이 사용해서 크롤링하기 

```python
from bs4 import BeautifulSoup as bs
```

```python
###@@@@@@중요
soup = bs(driver.page_source,'lxml')
```

```python
main_list = soup.select('h4.h')

# or soup.find_all('h4',class_='h fz_03')    <--- 이렇게 해도 된다.
```

```python
list1 = []
for i in range(0,len(main_list)) :
    list1.append(main_list[i])
    print(main_list[i].text)
```

```python
print(list1) #--->list1 에 담김.
```

```python
for i in list1 :
    print(i.text)
```

###  예외처리 (try문에서 에러가 나면 except로 빠짐)

```python
#더보기 선택
btnMore = driver.find_element_by_class_name('c_05')
#################----??????????????
while True :
    try btnMore.click()
        time.sleep(2)
    except ValueError :
        print('더보기 없음')
        break
```

```python
#더보기 선택
btnMore = driver.find_element_by_class_name('c_05') #-->왜 a.c_05 하면 안되지?
try :
    
    for index in range(500) :
        btnMore.click()
        time.sleep(2)
except:
    print(index)
```

```python
soup = bs(driver.page_source,'lxml')
main_list2 = soup.select('h4.h')
list2 = []
for i in range(0,len(main_list2)) :
    list2.append(main_list2[i].text)
    print(main_list2[i].text)
```

```python
main_list3 = soup.select('div.item-price>strong')
list3 = []
for i in range(0,len(main_list3)) :
    list3.append(main_list3[i].text)
    print(main_list3[i].text)
```

```python
main_list3 = soup.select('div.item-price')
list4 = []
for i in range(1,len(main_list3)+1) :
    list4.append(i)
print(list4)
```

```python
import pandas as pd
price_info = {'rank' : list4, 'name' : list2, 'price' : list3}
```

```python
price_info
```

```python
pd.DataFrame(price_info)
```

```python
Dataframe=pd.DataFrame(price_info)
Dataframe.set_index('rank', inplace = True )
```

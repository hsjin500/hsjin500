# 2020년 6월 1일 ~ 2020년 6월 20일 까지 가져오기

```python
M_day = []
M_rank = []
M_title = []
M_rate = []

for day in range(601,621,1) :
    naver_rank_rate_Url2 = f'https://movie.naver.com/movie/sdb/rank/rmovie.nhn?sel=cur&date=20200{day}'
    res = req.get(naver_rank_rate_Url2)

    list2 = bs(res.text,'lxml')
    M_title2 = list2.select('div.tit5>a')
    M_rate2 =list2.select('td.point')
    
    for i in range(len(M_title2)) :
        M_day.append(day)
        M_rank.append(i+1)
        M_title.append(M_title2[i].text)
        M_rate.append(M_rate2[i].text)
```

```python
M_info = {'date' : M_day, 'rank' : M_rank, 'title' : M_title, 'rate' : M_rate }
M_info
```

```python
data = pd.DataFrame(M_info)
```

```python
data.set_index('date',inplace = True)
data
```

### pandas를 활용한 날짜 생성 

```python
date = pd.date_range(start = ' 2019-01-01', end = '2020-06-20')
date
```

```python
days = date.strftime('%Y%m%d')
days
```

```python
M_day = []
M_rank = []
M_title = []
M_rate = []

for day in tqdm_notebook(days) :
    naver_rank_rate_Url2 = f'https://movie.naver.com/movie/sdb/rank/rmovie.nhn?sel=cur&date={day}'
    res = req.get(naver_rank_rate_Url2)

    list2 = bs(res.text,'lxml')
    M_title2 = list2.select('div.tit5>a')
    M_rate2 =list2.select('td.point')
    
    for i in range(len(M_title2)) :
        M_day.append(day)
        M_rank.append(i+1)
        M_title.append(M_title2[i].text)
        M_rate.append(M_rate2[i].text)
```

```python
M_info = {'date' : M_day, 'title' : M_title, 'rate' : M_rate}
M_info
```

```python
data1 = pd.DataFrame(M_info)
data1.set_index('date', inplace = True)
data1
```

```python
movie_review_Url = f'https://movie.naver.com/movie/bi/mi/pointWriteFormList.nhn?code=189537&type=after&isActualPointWriteExecute=false&isMileageSubscriptionAlready=false&isMileageSubscriptionReject=false&page=3'
h = {'user-agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/83.0.4103.106 Safari/537.36'}
res = req.get(movie_review_Url, headers = h)

res = bs(res.text,'lxml')
```

```python
Re_view = res.select('div.score_reple > p')
len(Re_view)
```

```python
res.select('div.score_reple > p')  #--->전체 댓글(아이콘 감상평 스포 포함)
```

```python
icon = res.select('span.ico_viewer') #---->관람객 아이콘 제거
for i in icon :
    i.extract()
```

```python
more = res.select('a.btn_more')   #---->감상평 제거
for m in more :
    m.extract()
```

```python
Spo = res.select('span.text_spo')  #---->스포일러 제거
for s in Spo :
    s.extract()
```

```python
page = res.select_one('strong.total > em') #---->page구문 하나의 리스트 틀 제거를 위해서  for문을 돌릴 필요가 없으니까 _one 사용
page = page.text

page = page.replace(',','')

page = int(page)

if page%10 > 0 :
    page = page//10 +1 
else :
    page = page//10


for page in range(page) :
    movie_review_Url = 'https://movie.naver.com/movie/bi/mi/pointWriteFormList.nhn?code=189537&type=after&isActualPointWriteExecute=false&isMileageSubscriptionAlready=false&isMileageSubscriptionReject=false&page='+page
    h = {'user-agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/83.0.4103.106 Safari/537.36'}
    res = req.get(movie_review_Url, headers = h)
    res = bs(res.text,'lxml')
    
    Re_view = res.select('div.score_reple > p')
    
    total_review = []
    
    icon = res.select('span.ico_viewer') #---->관람객 아이콘 제거
    for i in icon :
        i.extract()
    more = res.select('a.btn_more')   #---->감상평 제거
    for m in more :
        m.extract()
    Spo = res.select('span.text_spo')  #---->스포일러 제거
    for s in Spo :
        s.extract()
    review = []
    for i in Re_view :
        review.append(i)
        print(f'{i+1}',' : ',i.text.strip())
    
```

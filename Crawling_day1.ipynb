# Beautiful Soup 사용하기
  -- > 가져온 페이지 정보에서 내가 필요한 데이터만 추출해낼때 사용.

```python
from bs4 import BeautifulSoup as bs
import requests as req
```

```python
naver_url = 'https://www.naver.com/'
```

```python
res = req.get(naver_url)
```

```python
res.text
```

```python
naver_list = bs(res.text,'lxml') #파싱 방법 4가지중 1개로 가장 좋아서 굳이 다른 것을 사용할 필요가 없음
```

```python
naver_list.find_all('li', class_ = 'nav_item')
```

```python
#find_all = 가져온 데이터에서 필요한 정보들을 다 가져오겠다. ---> list형태로 데이터 반환
res = naver_list.find_all('li', class_ = 'nav_item')
```

```python
res[2]
```

```python
res[2].text
```

# 넘어온 데이터들의 모든 텍스트만 출력 

```python
for index in res :
    print(index.text)
```

```python
#필요없는 데이터를 지울때 사용한다.
for index in res :
    print(index.text.strip())
```

### 연령별 실시간 검색어 크롤링

```python
Age = input('검색 연령을 입력하세요 >> ')
realtime_naver_url = f'https://datalab.naver.com/keyword/realtimeList.naver?age={Age}s&where=main'
import requests as req
from bs4 import BeautifulSoup as bs
req.get(realtime_naver_url)
h = {'user-agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/83.0.4103.106 Safari/537.36'}
res = req.get(realtime_naver_url, headers = h)
realtime_list = bs(res.text, 'lxml')
res = realtime_list.find_all('span',class_='item_title')
#B = res.text.strip()
#for i in range(1,len(res)+1) :
#    print(f'{i}. {B}')
a= 0
for index in res :
    a+=1
    print("{}.".format(a),index.text.strip())
```

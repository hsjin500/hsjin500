## merge 함수

```python
import pandas as pd
```

```python
df5 = pd.DataFrame({'key' : ['K0','K2','K3','K4'], 'A' : ['A0','A1','A2','A3'], 'B' : ['B0', 'B1','B2','B3']})
df6 = pd.DataFrame({'key' : ['K0','K1','K2','K3'], 'C' : ['C0','C1','C2','C3'], 'D' : ['D0', 'D1','D2','D3']})
```

```python
df5
```

```python
df6
```

```python
pd.merge(df5,df6, on='key')
```

```python
from IPython.display import Image as image
image('C:/Users/SMT047/Desktop/수업/data/merge.png')
```

```python
pd.merge(df5,df6, how='left',on='key')
# 왼쪽 데이터를 기준으로 합치되 NaN이 있는 행은 제일 아래로 이동한다.
```

```python
pd.merge(df5,df6, how='right',on='key')
# 오른쪽 데이터를 기준으로 합치되 NaN이 있는 행은 제일 아래로 이동한다.
```

```python
pd.merge(df5,df6, how='outer',on='key') #합집합
```

```python
pd.merge(df5,df6, how='inner',on='key') #교집합
```

# matplotlib  : 데이터 시각화의 또다른 방법

```python
import matplotlib.pyplot as plt
%matplotlib inline 
#%matplotlib inline : 주피터 노트북을 실행한 브라우저 내부(inline)에서 그림이 바로 보일 수 있게 설정하는 것.
```

#### line plot

```python
y = [2,4,6]
plt.plot(y)
```

```python
plt.show()
#주피터 노트북말고 다른 일반적인 환경에서는 show를 해주어야 차트가 보인다.
```

```python
x=[1,2,3]
plt.plot(x,y)
plt.show()
```

```python
plt.plot(x,y,ls='-.')
plt.show()
```

```python
plt.plot(x,y,ls='1')
```

```python
plt.plot(x,y,marker='D',ls='--',lw=2.2, color='m',markerfacecolor='w',markersize=10)
plt.show()
```
